## Sickle Cell Disease: Unmet Medical Need

:::: {.columns}

::: {.column width="60%"}
<div class="feature-card">

**Disease Background**

- Single nucleotide mutation in β-globin gene (HBB)
- Polymerization of deoxygenated sickle hemoglobin (HbS)
- Vaso-occlusive crises, chronic hemolysis, organ damage
- ~100,000 patients in US; millions worldwide

</div>

<div class="highlight-box" style="margin-top: 1em;">

**Therapeutic Hypothesis**

Induction of fetal hemoglobin (HbF/γ-globin) prevents HbS polymerization and ameliorates disease phenotype

</div>
:::

::: {.column width="40%"}
<div class="status-card">

**Current Therapies**

- Hydroxyurea (HbF inducer)
- Voxelotor (anti-sickling)
- Crizanlizumab (anti-adhesion)
- Gene therapies (recent approvals)

**Opportunity:** Small molecule HbF inducers with improved efficacy and safety profiles

</div>
:::

## Primary Erythroid Culture Development

:::: {.columns}

::: {.column width="50%"}
<div class="status-card">

**Culture Protocol Optimization**

**Phase 1: Expansion (Days 0-7)**
- CD34+ cell isolation from mobilized peripheral blood
- Cytokine cocktail: SCF, IL-3, EPO
- Cell density optimization
- Media formulation testing

**Phase 2: Differentiation (Days 7-14)**
- Erythroid commitment
- Hemoglobin synthesis
- Terminal differentiation markers

</div>
:::

::: {.column width="50%"}
<div class="feature-card">

**Key Optimizations**

- **Donor Selection:** 3-5 SCD donors qualified
- **Seeding Density:** 5×10⁴ cells/mL optimal
- **Compound Addition:** Day 7 timing critical
- **Culture Vessel:** 96-well deep well plates
- **Incubation Time:** 7-day post-compound treatment

</div>

<div class="highlight-box" style="margin-top: 1em;">

**Quality Control Metrics**

Cell viability >80% | Enucleation rate | CD71/CD235a expression profile

</div>
:::

## Nanostring nCounter Implementation

:::: {.columns}

::: {.column width="55%"}
<div class="merck-blue-box">

**Technology Overview**

Digital molecular barcoding for direct mRNA quantification without amplification

</div>

<div class="feature-card" style="margin-top: 1em;">

**Custom CodeSet Design**

- **Target Genes:** HBG1, HBG2 (γ-globin), HBB, HBA1/2
- **Hemoglobin Switching:** BCL11A, KLF1, MYB, SOX6
- **Erythroid Markers:** GATA1, TAL1, EPOR
- **Housekeeping:** GAPDH, ACTB, HPRT1
- **Total:** 15-20 genes per panel

</div>
:::

::: {.column width="45%"}
<div class="status-card">

**Workflow Implementation**

1. Cell lysis & mRNA isolation
2. Hybridization with probe sets (16-18 hrs)
3. Nanostring nCounter detection
4. Digital counting & normalization
5. Data analysis pipeline

**Throughput:** 12-24 samples per run

**Turnaround:** 48 hours

</div>

<div class="highlight-box" style="margin-top: 1em;">

**Key Advantage:** Minimal sample input (100 ng RNA)

</div>
:::

## Assay Validation: Tool Compounds

:::: {.columns}

::: {.column width="50%"}
<div class="merck-blue-box">

**Positive Controls**

</div>

<div class="feature-card" style="margin-top: 1em;">

**Pomalidomide**
- Cereblon modulator
- Known HbF inducer
- EC₅₀: ~50-100 nM (Nanostring)
- Robust dose-response

**Hydroxyurea**
- Clinical standard
- Moderate HbF induction
- Higher concentrations required
- Benchmark comparator

</div>
:::

::: {.column width="50%"}
<div class="status-card">

**Validation Results**

**mRNA-Protein Correlation:**
- R² = 0.87 between HBG expression and HbF protein
- Dose-response concordance across platforms

**Reproducibility:**
- Inter-assay CV: 12-18%
- Intra-assay CV: 8-12%
- Donor-to-donor variability managed

**Selectivity:**
- No effect on HBB expression at active concentrations
- Erythroid differentiation maintained

</div>
:::

## Integration into Lead Optimization Workflow

:::: {.columns}

::: {.column width="40%"}
<div class="status-card">

**Screening Cascade**

1. **Primary Screen**
   - alphaLISA (HbF protein)
   - Single concentration or dose-response
   - Throughput: ~500 compounds/month

2. **Secondary Confirmation**
   - Nanostring gene panel
   - Mechanism insights
   - Multiple donors

3. **Profiling**
   - Full gene expression analysis
   - Off-target assessment
   - Cytotoxicity counter-screen

</div>
:::

::: {.column width="60%"}
<div class="feature-card">

**Decision-Making Impact**

**SAR Development:**
- Identified 3 chemical series with EC₅₀ <100 nM
- Guided optimization of potency and selectivity
- Eliminated compounds with erythroid toxicity

**Mechanistic Insights:**
- BCL11A downregulation pathway confirmed
- Novel KLF1-independent inducers discovered
- Chromatin modifier class characterized

**Translational Path:**
- Primary cell data informed candidate selection
- Donor variability predicted clinical heterogeneity
- De-risked progression decisions

</div>

## Challenges & Solutions

:::: {.columns}

::: {.column width="50%"}
<div class="feature-card">

**Technical Challenges**

**1. Donor Variability**
- **Issue:** 3-5 fold baseline HbF differences
- **Solution:** Normalized fold-change metrics; qualified donor panel

**2. Culture Reproducibility**
- **Issue:** Batch-to-batch media variation
- **Solution:** Lot testing; controlled serum batches

**3. Throughput Limitations**
- **Issue:** Manual culture steps
- **Solution:** Semi-automation with Hamilton; batch processing

</div>
:::

::: {.column width="50%"}
<div class="feature-card">

**Analytical Challenges**

**4. Nanostring Sample Throughput**
- **Issue:** 12-24 samples/run limitation
- **Solution:** Prioritized compounds; alphaLISA primary screen

**5. Data Integration**
- **Issue:** Correlating mRNA and protein data
- **Solution:** Custom analysis pipeline; unified database

**6. Assay Interference**
- **Issue:** Autofluorescence from compounds
- **Solution:** AlphaLISA format immune to interference

</div>
:::
